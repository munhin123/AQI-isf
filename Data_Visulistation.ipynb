```python
aqidata = {'Average': ['38.204795', '64.514992', '76.934502', '35.325127', '31.126332', '28.822216'],
          'Year': ['2018', '2019', '2020', '2021', '2022', '2023']}
aqiavg = pd.DataFrame(aqidata)
aqiavg['Average'] = aqiavg['Average'].astype(float)
aqiavg
```

Below is a Bar graph

```python
fig, ax = plt.subplots()
x = list(aqiavg['Year'])
y = list(aqiavg['Average'])
ax.set_title('Average AQI per year ACT')
ax.bar(x, y, width=1, edgecolor="white", linewidth=0.7)
ax.set_xlabel('Year')
ax.set_ylabel('AQI')

plt.show()
```

Below is line graph

```python
fig, ax = plt.subplots()
x = list(aqiavg['Year'])
y = list(aqiavg['Average'])
ax.plot(x, y, linewidth=2.0)
ax.set_title('Average AQI per year ACT')
ax.set_xlabel('Year')
ax.set_ylabel('AQI')


plt.show()
```

Below is Scatter Plot

```python
fig, ax = plt.subplots()
x = list(aqiavg['Year'])
y = list(aqiavg['Average'])
ax.scatter(x, y)
ax.set_title('Average AQI per year ACT')
ax.set_xlabel('Year')
ax.set_ylabel('AQI')

plt.show()
```

```python
civic = aqi2018[(aqi2018.Date == "15-Jan-18")&(aqi2018.Name == 'Civic')]
civic['Time'] = pd.to_datetime(civic['Time'], format='%H:%M')
civic = civic.sort_values(by='Time')
civic['Time'] = civic['Time'].dt.time
civic.loc[:, ~civic.columns.isin(['GPS', 'Date', 'NO2', 'O3_1hr', 'O3_4hr', 'O3_8hr', 'CO', 'PM10 1 hr', 'PM2.5 1 hr', 'PM10', 'PM2.5', 'AQI_CO', 'AQI_NO2', 'AQI_O3_1hr', 'AQI_O3_4hr', 'AQI_O3_8hr', 'AQI_PM10', 'AQI_PM2.5', 'Name', 'DateTime' ])]
```

```python
fig, ax = plt.subplots()
civic['Time'] = civic['Time'].astype(str)
x = list(civic['Time'])
y = list(civic['AQI_Site'])
ax.plot(x, y, linewidth=2.0)
plt.xticks(rotation=90)
ax.set_title('AQI Over A Day: Civic 15-Jan-2018')
ax.set_xlabel('Time')
ax.set_ylabel('AQI')


plt.show()
```
